```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
movie = pd.read_csv("C:/Users/DELL/Downloads/movie_dataset.csv",delimiter=",",low_memory=False)
movie.head()
```

```python
def get_title_from_index(index):
    return movie[movie.index == index]["title"].values[0]
def get_index_from_title(title):
    return movie[movie.title == title]["index"].values[0]
```

```python
features = ['keywords','cast','genres','director']
def combination(row):
    return row['keywords']+" "+row['cast']+" "+row['genres']+" "+row['director']
for i in features:
    movie[i] = movie[i].fillna('') 

movie["combination"] = movie.apply(combination,axis=1)
movie.iloc[0].combination
```

```python
cv=TfidfVectorizer()
tfidf_matrix=cv.fit_transform(movie['combination'])
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
indices=pd.Series(movie.index,index=movie['title'])
titles=movie['title']
def recommendations(title):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:20]
    movie_indices = [i[0] for i in sim_scores]
    return titles.iloc[movie_indices]
```

```python
recommendations('Avengers: Age of Ultron')
```
